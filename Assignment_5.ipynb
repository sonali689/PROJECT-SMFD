```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
import seaborn as sns
from tqdm import tqdm
import warnings
warnings.filterwarnings('ignore')

# Set random seed for reproducibility
np.random.seed(42)

# Configure plotting style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

print("LOOKBACK CALL OPTION PRICING")
print("Monte Carlo Simulation with Euler-Maruyama Method")

# PARAMETERS AND ASSUMPTIONS

# Market Parameters
S0 = 100.0      # Initial stock price
T = 1.0         # Time to maturity (1 year)
r = 0.05        # Risk-free interest rate (5%)
sigma = 0.2     # Volatility (20%)

# Simulation Parameters
N_simulations = 100000  # Number of Monte Carlo simulations
N_steps = 252          # Number of time steps (daily steps for 1 year)
dt = T / N_steps       # Time step size

print("ASSUMPTIONS:")
print("1. Geometric Brownian Motion (GBM) for underlying asset")
print("2. Constant volatility and risk-free rate")
print("3. No dividends or transaction costs")
print("4. Continuous monitoring of asset price")
print("5. European-style exercise (only at maturity)")
print()

print(f"Market Parameters:")
print(f"Initial Stock Price (S0): ${S0}")
print(f"Time to Maturity (T): {T} years")
print(f"Risk-free Rate (r): {r*100}%")
print(f"Volatility (σ): {sigma*100}%")
print(f"\nSimulation Parameters:")
print(f"Number of Simulations: {N_simulations:,}")
print(f"Number of Time Steps: {N_steps}")
print(f"Time Step Size (dt): {dt:.6f}")
print()

# GEOMETRIC BROWNIAN MOTION SIMULATION

def simulate_gbm_paths(S0, r, sigma, T, N_steps, N_simulations):
    """
    The SDE for GBM is: dS_t = r*S_t*dt + σ*S_t*dW_t
    Euler-Maruyama discretization: S_{t+1} = S_t + r*S_t*dt + σ*S_t*sqrt(dt)*Z
    where Z ~ N(0,1)
    
    Parameters:
    S0 : float - Initial stock price
    r : float - Risk-free rate
    sigma : float - Volatility
    T : float - Time to maturity
    N_steps : int - Number of time steps
    N_simulations : int - Number of simulation paths
    
    Returns:
    numpy.ndarray - Shape (N_simulations, N_steps+1) containing all price paths
    """
    dt = T / N_steps
    
    # Initialize price matrix
    S = np.zeros((N_simulations, N_steps + 1))
    S[:, 0] = S0  # Set initial prices
    
    # Generate random numbers for all paths at once (vectorized)
    Z = np.random.standard_normal((N_simulations, N_steps))
    
    # Euler-Maruyama simulation
    print("Simulating asset price paths using Euler-Maruyama method...")
    for i in tqdm(range(N_steps), desc="Time steps"):
        drift = r * S[:, i] * dt
        diffusion = sigma * S[:, i] * np.sqrt(dt) * Z[:, i]
        S[:, i + 1] = S[:, i] + drift + diffusion
        
        # positive prices (numerical stability)
        S[:, i + 1] = np.maximum(S[:, i + 1], 0.001)
    
    return S

# asset price paths
asset_paths = simulate_gbm_paths(S0, r, sigma, T, N_steps, N_simulations)

# Create time grid
time_grid = np.linspace(0, T, N_steps + 1)

print(f" generated {N_simulations:,} asset price paths")
print(f"  Final prices range: ${asset_paths[:, -1].min():.2f} - ${asset_paths[:, -1].max():.2f}")
print(f"  Average final price: ${asset_paths[:, -1].mean():.2f}")
print()

#  LOOKBACK CALL OPTION PRICING

def price_lookback_call_floating(asset_paths, r, T):
    """
    Price Lookback Call Option with floating strike using Monte Carlo simulation.
    
    A lookback call option with floating strike has payoff:
    Payoff = S_T - min(S_t for t in [0,T])
    
    The "strike" is the minimum price observed during the option's life.
    This guarantees a positive payoff since S_T ≥ min(S_t).
    
    Parameters:
    asset_paths : numpy.ndarray - Price paths (N_simulations × N_steps+1)
    r : float - Risk-free rate
    T : float - Time to maturity
    
    Returns:
    tuple: (option_price, payoffs, final_prices, min_prices)
    """
    final_prices = asset_paths[:, -1]
    min_prices = np.min(asset_paths, axis=1)
    
    # Calculate payoffs
    payoffs = final_prices - min_prices
    
    # Discount to present value
    option_price = np.exp(-r * T) * np.mean(payoffs)
    
    return option_price, payoffs, final_prices, min_prices

print("LOOKBACK CALL OPTION (FLOATING STRIKE)")
print("Payoff Formula: S_T - min(S_t for t ∈ [0,T])")
print("Where:")
print("- S_T: Final stock price at maturity")
print("- min(S_t): Minimum price observed during option life")
print("- This guarantees positive payoff since S_T ≥ min(S_t)")
print()

# Price the lookback call option
print("Calculating lookback call option price...")
lookback_price, lookback_payoffs, final_prices, min_prices = price_lookback_call_floating(asset_paths, r, T)

print(f" Lookback Call Option Price: ${lookback_price:.4f}")
print()

# STATISTICAL ANALYSIS

print("STATISTICAL ANALYSIS")

# statistics
mean_payoff = np.mean(lookback_payoffs)
std_payoff = np.std(lookback_payoffs)
min_payoff = np.min(lookback_payoffs)
max_payoff = np.max(lookback_payoffs)

mean_final = np.mean(final_prices)
mean_min = np.mean(min_prices)
mean_drawdown = mean_final - mean_min

print(f"Payoff Statistics:")
print(f"  Mean Payoff: ${mean_payoff:.4f}")
print(f"  Std Dev: ${std_payoff:.4f}")
print(f"  Min Payoff: ${min_payoff:.4f}")
print(f"  Max Payoff: ${max_payoff:.4f}")
print()

print(f"Price Statistics:")
print(f"  Mean Final Price: ${mean_final:.4f}")
print(f"  Mean Minimum Price: ${mean_min:.4f}")
print(f"  Mean Drawdown: ${mean_drawdown:.4f}")
print()

# confidence intervals
def calculate_confidence_intervals(payoffs, confidence_level=0.95):
    """Calculate confidence intervals for option prices."""
    mean_payoff = np.mean(payoffs)
    std_error = np.std(payoffs) / np.sqrt(len(payoffs))
    
    # Z-score for given confidence level
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)
    
    lower_bound = mean_payoff - z_score * std_error
    upper_bound = mean_payoff + z_score * std_error
    
    return lower_bound, upper_bound

lower_ci, upper_ci = calculate_confidence_intervals(lookback_payoffs)
discounted_lower = np.exp(-r * T) * lower_ci
discounted_upper = np.exp(-r * T) * upper_ci

print(f"95% Confidence Interval for Option Price:")
print(f"  Lower Bound: ${discounted_lower:.4f}")
print(f"  Upper Bound: ${discounted_upper:.4f}")
print()

# PLOTS
 
# Create plots
fig = plt.figure(figsize=(16, 12))

# Plot 1: Sample price paths with minimum tracking
ax1 = plt.subplot(2, 3, 1)
n_sample_paths = 50
for i in range(n_sample_paths):
    plt.plot(time_grid, asset_paths[i, :], alpha=0.3, linewidth=0.8, color='blue')
    # Highlight minimum price point
    min_idx = np.argmin(asset_paths[i, :])
    plt.plot(time_grid[min_idx], asset_paths[i, min_idx], 'ro', markersize=4, alpha=0.7)

plt.xlabel('Time (years)')
plt.ylabel('Stock Price ($)')
plt.title(f'Sample Asset Price Paths (First {n_sample_paths} paths)\nRed dots show minimum prices')
plt.grid(True, alpha=0.3)

# Plot 2: Distribution of final prices
ax2 = plt.subplot(2, 3, 2)
plt.hist(final_prices, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(x=mean_final, color='red', linestyle='--', linewidth=2, 
            label=f'Mean = ${mean_final:.2f}')
plt.xlabel('Final Stock Price ($)')
plt.ylabel('Density')
plt.title('Distribution of Final Stock Prices')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 3: Distribution of minimum prices
ax3 = plt.subplot(2, 3, 3)
plt.hist(min_prices, bins=50, density=True, alpha=0.7, color='lightcoral', edgecolor='black')
plt.axvline(x=mean_min, color='red', linestyle='--', linewidth=2, 
            label=f'Mean = ${mean_min:.2f}')
plt.xlabel('Minimum Stock Price ($)')
plt.ylabel('Density')
plt.title('Distribution of Minimum Stock Prices')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 4: Payoff distribution
ax4 = plt.subplot(2, 3, 4)
plt.hist(lookback_payoffs, bins=50, density=True, alpha=0.7, color='lightgreen', edgecolor='black')
plt.axvline(x=mean_payoff, color='red', linestyle='--', linewidth=2, 
            label=f'Mean = ${mean_payoff:.2f}')
plt.xlabel('Payoff ($)')
plt.ylabel('Density')
plt.title('Lookback Call Option Payoff Distribution')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 5: Scatter plot of Final vs Minimum prices
ax5 = plt.subplot(2, 3, 5)
plt.scatter(min_prices, final_prices, alpha=0.1, s=1)
plt.plot([min_prices.min(), min_prices.max()], [min_prices.min(), min_prices.max()], 
         'r--', linewidth=2, label='Final = Minimum')
plt.xlabel('Minimum Price ($)')
plt.ylabel('Final Price ($)')
plt.title('Final Price vs Minimum Price')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 6: Relationship between drawdown and payoff
ax6 = plt.subplot(2, 3, 6)
drawdowns = final_prices - min_prices
plt.scatter(drawdowns, lookback_payoffs, alpha=0.1, s=1)
plt.plot([drawdowns.min(), drawdowns.max()], [drawdowns.min(), drawdowns.max()], 
         'r--', linewidth=2, label='Perfect correlation')
plt.xlabel('Drawdown (Final - Min) ($)')
plt.ylabel('Payoff ($)')
plt.title('Drawdown vs Payoff (Should be identical)')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# CONVERGENCE ANALYSIS

def convergence_analysis(asset_paths):
    sample_sizes = np.logspace(2, np.log10(N_simulations), 20).astype(int)
    prices = []
    
    print("Performing convergence analysis...")
    for n in tqdm(sample_sizes, desc="Sample sizes"):
        price, _, _, _ = price_lookback_call_floating(asset_paths[:n], r, T)
        prices.append(price)
    
    return sample_sizes, prices

# Convergence analysis
sample_sizes, convergence_prices = convergence_analysis(asset_paths)

plt.figure(figsize=(10, 6))
plt.semilogx(sample_sizes, convergence_prices, 'b-', linewidth=2, marker='o', markersize=4)
plt.axhline(y=lookback_price, color='red', linestyle='--', linewidth=2, 
            label=f'Final Estimate = ${lookback_price:.4f}')
plt.xlabel('Number of Simulations')
plt.ylabel('Option Price ($)')
plt.title('Convergence of Monte Carlo Estimate')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# SENSITIVITY ANALYSIS

print("SENSITIVITY ANALYSIS")
print("="*50)

def sensitivity_analysis():
    """Analyze sensitivity to volatility and interest rate."""
    
    # Volatility sensitivity
    print("Analyzing volatility sensitivity...")
    sigmas = np.linspace(0.1, 0.4, 10)
    prices_vol = []
    
    for vol in tqdm(sigmas, desc="Volatility levels"):
        paths = simulate_gbm_paths(S0, r, vol, T, N_steps, 10000)  # Reduced for speed
        price, _, _, _ = price_lookback_call_floating(paths, r, T)
        prices_vol.append(price)
    
    # Interest rate sensitivity
    print("Analyzing interest rate sensitivity...")
    rates = np.linspace(0.01, 0.1, 10)
    prices_rate = []
    
    for rate in tqdm(rates, desc="Interest rate levels"):
        paths = simulate_gbm_paths(S0, rate, sigma, T, N_steps, 10000)
        price, _, _, _ = price_lookback_call_floating(paths, rate, T)
        prices_rate.append(price)
    
    return sigmas, prices_vol, rates, prices_rate

sigmas, prices_vol, rates, prices_rate = sensitivity_analysis()

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(sigmas * 100, prices_vol, 'b-', linewidth=2, marker='o')
plt.axvline(x=sigma * 100, color='red', linestyle='--', linewidth=2, 
            label=f'Base Case = {sigma*100}%')
plt.xlabel('Volatility (%)')
plt.ylabel('Lookback Call Price ($)')
plt.title('Sensitivity to Volatility')
plt.legend()
plt.grid(True, alpha=0.3)

plt.subplot(1, 2, 2)
plt.plot(rates * 100, prices_rate, 'r-', linewidth=2, marker='o')
plt.axvline(x=r * 100, color='blue', linestyle='--', linewidth=2, 
            label=f'Base Case = {r*100}%')
plt.xlabel('Risk-free Rate (%)')
plt.ylabel('Lookback Call Price ($)')
plt.title('Sensitivity to Interest Rate')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# FINAL SUMMARY

print("FINAL RESULTS SUMMARY")

print(f"LOOKBACK CALL OPTION (FLOATING STRIKE)")
print(f"Monte Carlo Price: ${lookback_price:.4f}")
print(f"95% Confidence Interval: [${discounted_lower:.4f}, ${discounted_upper:.4f}]")
print()

print(f"SIMULATION VALIDATION:")
print(f"- Number of paths: {N_simulations:,}")
print(f"- Time steps per path: {N_steps}")
print(f"- All payoffs positive: {np.all(lookback_payoffs >= 0)}")
print(f"- Perfect correlation between drawdown and payoff: {np.corrcoef(drawdowns, lookback_payoffs)[0,1]:.6f}")
print()

print(f"KEY CHARACTERISTICS:")
print(f"- Guaranteed positive payoff (S_T ≥ min(S_t))")
print(f"- Path-dependent (depends on entire price history)")
print(f"- Higher value than European options due to hindsight advantage")
print(f"- Sensitive to volatility (higher vol = higher option value)")
print()

print(f"PRACTICAL APPLICATIONS:")
print(f"- Portfolio protection with guaranteed upside")
print(f"- Compensation structures (executives, fund managers)")
print(f"- Insurance products with market participation")
print(f"- Structured products for conservative investors")
```
